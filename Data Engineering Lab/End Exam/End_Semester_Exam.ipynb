```python
words = gutenberg.words(fileids='chesterton-ball.txt')
words = [word for word in words if word.isalnum()]       #remove nonalphanumeric
words = [word.lower() for word in words] 
freq_dist = FreqDist(words) 
```

```python
common2 = freq_dist.most_common(2)  
print('2 frequent words and count\n')
for item in common2:
    print(f"{item[0]}: {item[1]}")
```

```python
least2 = freq_dist.most_common()[-2:]
print('2 least frequent words and count\n')
for item in least2:
    print(f"{item[0]}: {item[1]}")
```

```python
words_new = []
for word in words:
    if word == 'the':
        words_new.append('11')
    elif word == 'and':
        words_new.append('ver')
    elif word == 'Ver':
        words_new.append('and')
    elif word == '11':
        words_new.append('the')
    else:
        words_new.append(word)
    
    
```

```python
freq_dist = FreqDist(words_new) 
common5 = freq_dist.most_common(5)  
print('5 frequent words and count\n')
for item in common5:
    print(f"{item[0]}: {item[1]}")
```

```python
features = df.columns
features 
```

```python
#Handling missing values 
```

```python
# Remove all columns which contain more than 25% missing values. Display the eliminated column names
tot_samples = len(df)
for feature in features:
    if len(df[df[feature]== 0]) > tot_samples/4:
        print('No. zeros in {} is {}, so dropped'.format(feature,len(df[df[feature]== 0])))
        df.drop([feature], axis=1, inplace=True)
        
print('\nShape of cleaned dataframe is {}'.format(df.shape))
```

```python
#Impute the missing values with either previous value , next value or mean value.

from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit_transform(df)
df
```

```python
# Anomaly detection and removal
# box plot
df.boxplot()
```

```python
features_new = df.columns
```

```python
# histogram
#histogram of all the columns with numeric values seperately
for feature in features_new:
    df.hist(column=feature)
```

```python
# from the box plot its clear that price has outlier
# removing the outlier
df.drop(df[df['price'] > 0.5e7].index, inplace=True)  
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

data = scaler.fit_transform(df[:-1])              # normalise escept price column 
```

```python
# modeling
from sklearn.model_selection import train_test_split
df = df.fillna(0)
X = df.iloc[:,:-1] #input
y = df.iloc[:,-1]  #output
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train, y_train
```

```python
#implementing linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train) 
```

```python
#Evaluation 
from sklearn.metrics import mean_squared_error
y_pred = model.predict(X_test)
MSE = mean_squared_error(y_pred,y_test)
print(f'MSE for test data:{MSE}')
```
