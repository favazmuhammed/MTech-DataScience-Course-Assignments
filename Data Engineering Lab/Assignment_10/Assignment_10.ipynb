###### 1. Write a Python program to display a pie chart for the top 10 most popular programming languages with the most popular wedge and the least popular wedge exploded. 

```python
prog_language=pd.read_csv('Programming_Language_Popularity.csv')
prog_language.head()
```

```python
def get_index(data, key):
    
    if key == 'minimum':
        return data.index(min(data))
    elif key == 'maximum':
        return data.index(max(data))
        
```

```python
fig, ax = plt.subplots(figsize=(12,12))


language = list(prog_language['Language'].values)
popularity = list(prog_language['Popularity Index'].values)
explode = [0 for i in range(len(prog_language))]
explode[get_index(popularity,'minimum')] =0.2
explode[get_index(popularity,'maximum')] =0.2


wedges, texts=ax.pie(popularity, labels=language,explode=explode, rotatelabels = True, startangle = 30, shadow=True)
plt.setp(texts, size=10, color ='red')
ax.set_title('Programming Languages Popularity')
plt.show()
plt.close()
```

###### 2.Write a Python program to display a bar chart of the popularity of programming Languages sorted in the order of highest to lowest popularity. Use different color for each bar. Attach a text label above each bar displaying its popularity (float value).

```python
fig, ax = plt.subplots(figsize =(16, 9))

df_sorted = prog_language.sort_values('Popularity Index', ascending=False)
color = ['green','blue','red','orange','yellow','grey','magenta']

bars=ax.bar('Language','Popularity Index', data=df_sorted, color=color)
ax.set_title('Programming Languages Popularity')
plt.xticks(rotation = 45)
plt.xlabel('Language')
plt.ylabel('Popularity')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval+0.5, yval)
    
plt.show()
plt.close()
```

###### 3. Use dataset "Corona_Tweets.csv" for this question. Write a Python program to display word cloud for the Corona_Tweets dataset. Print the most popular (frequent) word.

```python
df = pd.read_csv('Corona_Tweets.csv')
df
```

```python
tweets = df['OriginalTweet'].str.replace('[^\w\s]','')
tweets
```

```python
text = " ".join(tweet.split()[1] for tweet in tweets)
text = text.lower()
```

```python
from wordcloud import WordCloud, STOPWORDS
wordcloud = WordCloud().generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
