```python
#import all necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, SCORERS
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

#import module for ignoring warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#read data from 'patients.csv' file
data = pd.read_csv('patients.csv')
data.head(10)
```

```python
Y=data['A44']                       #column 'A44' as output
X=data.drop(['A44'], axis=1)        #remaining columns as features
X.head(), Y.head()
```

```python
#import GridSearchCV for finding optimal alpha for lasso regression
from sklearn.model_selection import GridSearchCV

parameters = {'alpha': [0.1, 1, 10,15,20,100]}             #define different alpha values in the parameter module
kfold = KFold(n_splits=10, shuffle=True, random_state=1)   #initialize kfold cross validation with 10 splits
model= GridSearchCV(Lasso(), parameters,cv=kfold, refit=True, verbose = 3,n_jobs=-1) #initialize GridSearchCV with model as Lasso
model.fit(X,Y)                                                #fit X and Y using Grid search

print("Optimal alpha is: %0.2f"%(model.best_params_['alpha']))  #printing optimal alpha for Lasso with CV
print("Best score is: %0.02f"%(model.best_score_))              #Best score for given alpha
```

## Question-2

```python
from sklearn import datasets
whole_data=datasets.load_diabetes(return_X_y=False)      #from sklear datasets import diabetes data set
                                                         #whole_data holds data, target, feature_names etc as a dictionary
X=pd.DataFrame(whole_data['data'])                       #create a dataframe 'X' holds features
Y=pd.DataFrame(whole_data['target'])                     #creat a dataframe 'Y' holds target values
X.columns=whole_data['feature_names']                    #rename column names with feature names and target for X and Y
Y.columns=['target']                                     
X.head(), Y.head()
```

```python
X.shape, Y.shape               
```

#### a. Perform k-fold CV on Ridge and Lasso Regression where k=(5 and 10)

```python
models=[('lasso', Lasso(alpha=0.01)),('ridge', Ridge(alpha=0.01))]          #define an array holding lasso and ridge models with alpha=0.01
K_values=[5,10]                                                             #define an array for holding k values for kfold
print("Model     K     Accuracy     Standard_Deviation")

#iterate over different models and defferent Ks
#for each case fins mean squared error
#accuray is root mean squared error
#print both these for lasso and ridge for k=5,k=10
for name,model in models:
    print(name)
    for k in K_values:
        kfold = KFold(n_splits=k, shuffle=True,random_state=10)
        scores = cross_val_score(model, X, Y, cv=kfold,scoring='neg_mean_squared_error')
        print("          %d     %0.02f          %0.02f" %(k, np.sqrt(abs(scores.mean())), scores.std()))
        
```

#### b. Perform LOO CV for Lasso and Ridge Regression.


```python
from sklearn.model_selection import LeaveOneOut
LOO = LeaveOneOut()                                      #define leave one out
print('model     Accuracy      Standard_Deviation')

#iterate over lasso and redge
#accuray is root mean squared error
for name,model in models:
        scores = cross_val_score(model, X, Y, cv=LOO,scoring='neg_mean_squared_error')
        print("%s     %f     %f " % (name, np.sqrt(abs(scores.mean())), scores.std()))
```

#### c. Perform Grid-Search CV for Lasso and Ridge Regression for various values of alpha and show best value of parameter.

```python
models = [('lasso', Lasso()), ('ridge', Ridge())]        #define array for holdinf lasso and redge models 
```

```python
#iterate over lasso and redge 
for name, model in models:
    parameters = {'alpha': [1e-15, 0.1, 1, 10,15,20,100]}      #define parameters for grid search
    kfold = KFold(n_splits=10, shuffle=True, random_state=10)   #initialize kfold with k=10
    
    #inititialize grid search with cv=k fold or cv=LOO,un comment one
    my_model= GridSearchCV(model, parameters,cv=kfold, refit=True, verbose = 3,n_jobs=-1)
    #my_model= GridSearchCV(model, parameters,cv=LOO, refit=True, verbose = 3,n_jobs=-1)
    
    my_model.fit(X,Y)                                        #fit the model with X and Y using optimal alpha value
    print("Best value of alpha for %s:"%name)
    print(my_model.best_params_)                             #print best value of alpha  
    
```
