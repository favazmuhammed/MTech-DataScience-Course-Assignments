```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons,make_blobs
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import binarize
```

## 1) Gaussian NB

```python
# creating dataset
X, y = make_blobs(n_samples=100,centers=2,n_features=2,random_state=0)

# splitting into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# initializing the classifier
clf = GaussianNB()

# fitting the model over training dataset
clf.fit(X_train, y_train)

# plotting the decision regions
plot_decision_regions(X_train, y_train, clf=clf, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('GaussianNB_blob')
plt.show()
```

```python
print("confusion matrix train")
print(confusion_matrix(y_train, clf.predict(X_train)))
print("confusion matrix test")
print(confusion_matrix(y_test, clf.predict(X_test)))
print("accuracy scores train")
print(accuracy_score(y_train, clf.predict(X_train)))
print("accuracy scores test")
print(accuracy_score(y_test, clf.predict(X_test)))
```

```python
# plotting the decision regions
plot_decision_regions(X_test, y_test, clf=clf, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('GaussianNB_blob')
plt.show()
```

## 2) Bernoulli NB

```python
from sklearn.naive_bayes import BernoulliNB
X, y = make_blobs(n_samples=100,centers=2,n_features=2,random_state=0)
X_train, X_test, Y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
#binarize make feature value of each point binary based on threshold values above threshold are given 1 same & below 0
#for binary class 0,0 & 0,1 are assumed to class 0 and 1,0 & 1,1 are assigned to class 1
bnb = BernoulliNB(binarize=0.0)
bnb.fit(X_train, Y_train)
plot_decision_regions(X_train, y_train, clf=bnb, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('BernoulliNB_blob')
plt.show()
```

```python
print("confusion matrix train")
print(confusion_matrix(y_train, bnb.predict(X_train)))
print("confusion matrix test")
print(confusion_matrix(y_test, bnb.predict(X_test)))
print("accuracy scores train")
print(accuracy_score(y_train, bnb.predict(X_train)))
print("accuracy scores test")
print(accuracy_score(y_test, bnb.predict(X_test)))
```

## 3) Logistic Regression

```python
X, y = make_blobs(n_samples=100,centers=2,n_features=2,random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, clf=clf, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Logistic Regresion_blob')
plt.show()
```

```python
print("confusion matrix train")
print(confusion_matrix(y_train, clf.predict(X_train)))
print("confusion matrix test")
print(confusion_matrix(y_test, clf.predict(X_test)))
print("accuracy scores train")
print(accuracy_score(y_train, clf.predict(X_train)))
print("accuracy scores test")
print(accuracy_score(y_test, clf.predict(X_test)))
```

## 4) KNN

```python
X, y = make_blobs(n_samples=100,centers=2,n_features=2,random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
plot_decision_regions(X_train, y_train, clf=clf, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('KNN')
plt.show()
```

```python
print("confusion matrix train")
print(confusion_matrix(y_train, clf.predict(X_train)))
print("confusion matrix test")
print(confusion_matrix(y_test, clf.predict(X_test)))
print("accuracy scores train")
print(accuracy_score(y_train, clf.predict(X_train)))
print("accuracy scores test")
print(accuracy_score(y_test, clf.predict(X_test)))
```

## MultinomialNB

```python
# creating dataset
X, y = make_blobs(n_samples=100,centers=2,n_features=2,random_state=0)

# converting values to positive
for i in range(len(X)):
    for j in range(len(X[0])):
        X[i][j]=X[i][j]**2
# splitting into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# initializing the classifier
clf = MultinomialNB()

# fitting the model over training dataset
clf.fit(X_train, y_train)

# plotting the decision regions
plot_decision_regions(X_train, y_train, clf=clf, legend=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('GaussianNB_blob')
plt.show()
```

```python
print("confusion matrix train")
print(confusion_matrix(y_train, clf.predict(X_train)))
print("confusion matrix test")
print(confusion_matrix(y_test, clf.predict(X_test)))
print("accuracy scores train")
print(accuracy_score(y_train, clf.predict(X_train)))
print("accuracy scores test")
print(accuracy_score(y_test, clf.predict(X_test)))
```
