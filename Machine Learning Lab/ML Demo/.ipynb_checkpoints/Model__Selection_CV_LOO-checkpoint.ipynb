### Cross Validation
 Cross validation is technique involves reserving a particular sample of a dataset on which you do not train the model. Later, you test the model on this sample before finalizing the model.

 Here are the steps involved in a cross validation:


*   You Reserve a sample dataset.
*   Train the model using the remaining part of the dataset.

*   Use the reserve sample of the dataset test(validation) set. This will help you to know the effectiveness of model performance, if your maodel delivers a positive result on the validation data, go ahead with current model.



image.png

```python
# importing kFold Cross validation and checking it's score
start_time = time.time()
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

kf = KFold(n_splits=10,shuffle=True,random_state=2)

clf = svm.SVR(kernel='rbf', C=1e3, gamma=0.1)
scores = cross_val_score(clf, X, y, cv=kf,scoring='neg_mean_squared_error')
# clf.predict(X_test)
print("\n--- %s seconds for executing k fold---\n" % (time.time() - start_time))


start_time = time.time()
data_y, data_yhat = list(), list()
X=np.asarray(X)
y=np.asarray(y)
# kfold = KFold(n_splits=10, shuffle=True)
# enumerate splits
for train_ix, test_ix in kf.split(X):
	# get data
	train_X, test_X = X[train_ix], X[test_ix]
	train_y, test_y = y[train_ix], y[test_ix]
	# fit model
	model = svm.SVR()
	model.fit(train_X, train_y)
	# make predictions
	yhat = model.predict(test_X)
	# store
	data_y.extend(test_y)
	data_yhat.extend(yhat)
# evaluate the model
acc = metrics.mean_squared_error(data_y, data_yhat)
print('Accuracy: %.3f' % (acc))
print("\n--- %s seconds for training and predicting using SVM---\n" % (time.time() - start_time))
rmse = np.sqrt(acc)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
```

 ### Grid Search CV

 It is the process of performing hyperparameter tuning in order to determine the optimal values for a given model. As mentioned above, the performance of a model significantly depends on the value of hyperparameters.

Checking Parameters manually could take a considerable amount of time and resources and thus we use GridSearchCV to automate the tuning of hyperparameters.

In this we Predefine the dictionary of hyperparamters we want to test and pass it to function.

GridSearchCV tries all the combinations of the values passed in the dictionary and evaluates the model for each combination using the Cross-Validation method. Hence after using this function we get accuracy/loss for every combination of hyperparameters and we can choose the one with the best performance.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
param_grid = {'C': [0.1, 1, 10, 100],   
              'gamma':['scale', 'auto'],
              'kernel': ['linear','rbf','sigmoid']}  
             
start_time = time.time()
grid = GridSearchCV(svm.SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1) 
   
# fitting the model for grid search 
grid.fit(X_train, y_train) 
 
# print best parameter after tuning 
print(f'\nThe List of best Parameters from the given tested parameters are:{grid.best_params_}')
print("--- %s seconds ---" % (time.time() - start_time))
grid_predictions = grid.predict(X_test) 
print('\n\n')


```

### To check which of the Available models will suit best for required dataset 

```python
#importing different methods

from sklearn.linear_model import LogisticRegression, LinearRegression, Lasso
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
start_time = time.time()
alpha = 10
# defing base parameters and appending different models 
num_folds = 10
lasso = Lasso(alpha=0.01)
seed = 7
scoring = 'accuracy'
models = []
models.append(('LiR',LinearRegression()))
models.append(('LaS',lasso))
models.append(('LR', LogisticRegression()))
# models.append(('LDA', LinearDiscriminantAnalysis()))
# models.append(('KNN', KNeighborsClassifier()))
# models.append(('CART', DecisionTreeClassifier()))
# models.append(('NB', GaussianNB()))
models.append(('SVM', svm.SVR( kernel='linear',C= 10, gamma= 'scale',)))

results = []
names = []
print('model   Accuracy    Standard_Deviation')

# training for every model and testing training accuracy for every model
for name, model in models:
    kfold = KFold(n_splits=num_folds, shuffle=True,random_state=seed)
    scores = cross_val_score(model, X_train, y_train, cv=kfold,scoring='neg_mean_squared_error')
    results.append(scores)
    names.append(name)
    msg = "%s     %f     %f " % (name, np.sqrt(abs(scores.mean())), scores.std())
    print(msg)

print("--- %s seconds ---" % (time.time() - start_time))
```
