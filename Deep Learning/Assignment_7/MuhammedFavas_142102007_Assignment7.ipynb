### Build and train a model with Convolutional layer/layers to classify the images in the test dataset as either ‘happy’ or ‘neutral’.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Input, layers
from keras.callbacks import ModelCheckpoint
```

```python
# function for ploting loss and accuracy for validation and test dataset
def plot_graph(history):
  val_loss = history.history['val_loss']
  train_loss = history.history['loss']
  val_accuracy = history.history['val_accuracy']
  train_accuracy = history.history['accuracy']
  x = list(range(1,len(val_loss)+1))
  fig,(ax1,ax2) = plt.subplots(1,2,figsize=(15, 5))
  ax1.plot(x,val_accuracy,'r',label = "Validation Accuarcy")
  ax1.plot(x,train_accuracy,'b',label = "Train Accuracy")
  ax1.set_xlabel('Epoch')
  ax1.set_ylabel('Accuarcy')
  ax1.legend()
  ax1.grid()

  ax2.plot(x,val_loss,'r',label = "Validation loss")
  ax2.plot(x,train_loss,'b',label = "Train loss")
  ax2.set_xlabel('Epoch')
  ax2.set_ylabel('Binary Cross Entropy')
  ax2.legend()
  ax2.grid()
  plt.show()
```

```python
# import train data as train and validation with 0.3 split to v alidation
training_ds= tf.keras.preprocessing.image_dataset_from_directory(
    directory = train_path,
    labels="inferred",
    label_mode="binary",
    color_mode="grayscale",
    batch_size=64,
    image_size=(48, 48),
    shuffle=True,
    seed= 10,
    validation_split= 0.3,
    subset="training"
)
validation_ds= tf.keras.preprocessing.image_dataset_from_directory(
    directory = train_path,
    labels="inferred",
    label_mode="binary",
    color_mode="grayscale",
    batch_size=32,
    image_size=(48, 48),
    shuffle=True,
    seed= 10,
    validation_split= 0.3,
    subset="validation"
)
```

### Custom Model

```python
# built model with functional API
inputs = keras.Input(shape=(48,48,1))
x = inputs / 255                                                                                #rescale input to [0,1]
for i in range(3):
  x = layers.Conv2D(filters=64, kernel_size=3, strides=1, activation="relu", padding="same")(x) # conv2D+ReLU
  x = layers.Conv2D(filters=64, kernel_size=3, strides=1, activation="relu", padding="same")(x) # conv2D+ReLU
  x = layers.MaxPooling2D(2)(x)                                                                 # maxpooling

x = layers.Flatten()(x)
for j in range(2):
  x = layers.Dense(units=64, activation="relu")(x)                                                # dense layer + ReLU
  x = layers.BatchNormalization()(x)                                                              # batch normalization

outputs = layers.Dense(units=1, activation="sigmoid")(x)                                          # output layer

model = keras.Model(inputs, outputs)
model.summary()
```

```python
# compile the model and train and save best model by monitering validation accuracy
loss = tf.keras.losses.BinaryCrossentropy()
opt = keras.optimizers.Adam(learning_rate=0.01)
mc = ModelCheckpoint('/content/drive/MyDrive/Assignments/Best_models/binary_classifier.h5',
                     monitor='val_accuracy', mode='max',  verbose=1, save_best_only=True)

model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
history = model.fit(training_ds,validation_data = validation_ds,epochs = 30, verbose=1, callbacks = [mc])
```

```python
# plot loss and accuracy curves for training and validation 
plot_graph(history)
```

## Prediction for test data

```python
model = keras.models.load_model('/content/drive/MyDrive/Assignments/Best_models/binary_classifier.h5')  # load best model
test_data = np.load('/content/drive/MyDrive/datasets/DL/FER/test/test_data.npy')                        # load the data
predicted_class = model.predict(test_data)                        # predict
predicted_class[:10]                                           # first 10 predicted probabilities
```

```python
# put threshold as 0.5 
# if output probability >=0.5 it is neutral class else happy class (happy is class 0 neutral is class 1)
index_class = list()
for i in range(len(predicted_class)):
  if predicted_class[i] >= 0.5:
    index_class.append((i,'neutral'))
  else:
    index_class.append((i,'happy'))

# create a csv file and save the results
output = pd.DataFrame(index_class, columns=['Index', 'Output'])
output.head(10)
```
